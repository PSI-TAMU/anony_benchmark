```python
import os
import re
import pandas as pd
import nemo.collections.asr as nemo_asr
from nemo.collections.asr.metrics.wer import word_error_rate
asr = nemo_asr.models.EncDecCTCModelBPE.restore_from("stt_en_conformer_ctc_xlarge_v1.10.0/stt_en_conformer_ctc_xlarge.tar.gz")
```

```python
audio_dir = "/data/mtseng/VCTK-Corpus/wav48/p315"
out_dir = "/data/mtseng/VCTK-Corpus/txt/{}".format(os.path.basename(audio_dir))
os.makedirs(out_dir, exist_ok=True)
```

```python
import glob
import tqdm
for audio_path in tqdm.tqdm(sorted(glob.glob(os.path.join(audio_dir, "*.wav")))):
    out_path = os.path.join(out_dir, os.path.basename(audio_path).replace(".wav", ".txt"))
    if os.path.exists(out_path):
        continue
    result = asr.transcribe([audio_path])
    pred_text = result[0]
    with open(out_path, "w") as f:
        f.write(pred_text + "\n")
```
