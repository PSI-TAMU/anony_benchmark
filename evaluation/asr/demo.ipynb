```python
import re
import nemo.collections.asr as nemo_asr
from nemo.collections.asr.metrics.wer import word_error_rate
asr = nemo_asr.models.EncDecCTCModelBPE.restore_from("stt_en_conformer_ctc_xlarge_v1.10.0/stt_en_conformer_ctc_xlarge.tar.gz")
```

```python
audio_path = "/data/mtseng/L2Artic/all_data_for_ac_vc/BDL/wav/arctic_a0001.wav"
result = asr.transcribe([audio_path])
```

```python
def normalize(text):
    # lowercase
    text = text.lower()
    # remove punctuation (anything not word or whitespace)
    text = re.sub(r"[^\w\s]", "", text)
    return text
```

```python
gt_text = "Author of the danger trail Philip Steels etc"
pred_text = result[0]
print("Predicted Text: ", pred_text)
print("GT text: ", gt_text)
```

```python
wer = word_error_rate(hypotheses=[normalize(pred_text)], references=[normalize(gt_text)])
print("WER:", wer)
```
