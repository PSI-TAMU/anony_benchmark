```python
import requests
from pathlib import Path
```

```python
sample_audio_path = Path('../samples/SLT_arctic_a0043.wav')
sample_audio_path = str(sample_audio_path.resolve())
print(f"Sample audio path: {sample_audio_path}")
```

### NISQA-MOS

```python
response = requests.post('http://localhost:8000/predict_mos', json={"audio_path": sample_audio_path})
print(response.json())
```

### ASR

```python
response = requests.post('http://localhost:8000/predict_asr', json={"audio_path": sample_audio_path})
print(response.json())
```

### WavLM Speaker Embedding

```python
response = requests.post('http://localhost:8000/speaker_embedding', json={"audio_path": sample_audio_path})
print(response.json())
```
